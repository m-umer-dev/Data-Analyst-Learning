```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("✅ Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print(f"\nFirst 5 rows:")
print(df.head())
```

```python
print("=== DATASET EXPLORATION ===")
print(f"\n1. Shape: {df.shape}")
print(f"\n2. Columns: {df.columns.tolist()}")
print(f"\n3. Data Types:\n{df.dtypes}")
print(f"\n4. Missing Values:\n{df.isnull().sum()}")
print(f"\n5. Missing Percentage:\n"
      f"{(df.isnull().sum()/len(df)*100).round(2)}")
print(f"\n6. Statistics:\n{df.describe()}")
```

```python
print("=== Passenger Class Distribution ===")
print(df["Pclass"].value_counts())

print("\n=== Survival Distribution ===")
print(df["Survived"].value_counts())

print("\n=== Gender Distribution ===")
print(df["Sex"].value_counts())

print("\n=== Embarkation Port ===")
print(df["Embarked"].value_counts())
```

```python
total = len(df)
survived = df["Survived"].sum()
survival_rate = (survived / total) * 100
avg_age = df["Age"].mean()
avg_fare = df["Fare"].mean()

print("=== TITANIC QUICK INSIGHTS ===")
print(f"Total Passengers: {total}")
print(f"Survivors: {survived}")
print(f"Survival Rate: {survival_rate:.1f}%")
print(f"Average Age: {avg_age:.1f} years")
print(f"Average Fare: ${avg_fare:.2f}")
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Filter survivors only
survivors = df[df["Survived"] == 1]
print(f"Total Survivors: {len(survivors)}")

# Filter first class passengers
first_class = df[df["Pclass"] == 1]
print(f"First Class Passengers: {len(first_class)}")

# Filter female passengers
females = df[df["Sex"] == "female"]
print(f"Female Passengers: {len(females)}")
```

```python
# Female survivors
female_survivors = df[(df["Sex"] == "female") &
                       (df["Survived"] == 1)]
print(f"Female Survivors: {len(female_survivors)}")

# First class survivors
first_class_survivors = df[(df["Pclass"] == 1) &
                            (df["Survived"] == 1)]
print(f"First Class Survivors: {len(first_class_survivors)}")

# Children (under 18)
children = df[df["Age"] < 18]
print(f"Children on board: {len(children)}")
print(f"Children who survived: "
      f"{children['Survived'].sum()}")
```

```python
print("=== Survival Rate by Gender ===")
gender_survival = df.groupby("Sex")["Survived"].mean() * 100
print(gender_survival.round(1))

print("\n=== Survival Rate by Class ===")
class_survival = df.groupby("Pclass")["Survived"].mean() * 100
print(class_survival.round(1))

print("\n=== Average Fare by Class ===")
class_fare = df.groupby("Pclass")["Fare"].mean()
print(class_fare.round(2))

print("\n=== Average Age by Class ===")
class_age = df.groupby("Pclass")["Age"].mean()
print(class_age.round(1))
```

```python
# Add age category
df["Age_Group"] = df["Age"].apply(
    lambda x: "Child" if x < 18
    else "Adult" if x < 60
    else "Senior" if pd.notna(x)
    else "Unknown"
)

print("=== Survival Rate by Age Group ===")
age_survival = df.groupby("Age_Group")["Survived"].agg(
    ["sum", "count", "mean"]
).round(2)
age_survival.columns = ["Survivors", "Total", "Rate"]
age_survival["Rate"] = (age_survival["Rate"] * 100).round(1)
print(age_survival)

print("\n=== Key Titanic Insight ===")
female_rate = df[df["Sex"]=="female"]["Survived"].mean()*100
male_rate = df[df["Sex"]=="male"]["Survived"].mean()*100
print(f"Female survival rate: {female_rate:.1f}%")
print(f"Male survival rate: {male_rate:.1f}%")
print(f"Women were {female_rate/male_rate:.1f}x more likely to survive!")
```
